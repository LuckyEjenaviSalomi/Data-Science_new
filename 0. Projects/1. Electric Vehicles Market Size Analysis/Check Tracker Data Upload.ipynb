## Import Libraries

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

## Read Data

```python
data = pd.read_excel(filepath, sheet_name='Issues & Recommendations')
```

## Rename Header

```python
first_row = np.array(data.head(1).values).flatten()
headers = data.columns
```

```python
data_dict = {"header1": list(headers), "header2": list(first_row)}
```

```python
df_headers = pd.DataFrame(data_dict)
```

```python
df_headers.dropna(inplace=True)
```

```python
rename_cols = dict(zip(df_headers['header1'], df_headers['header2']))
```

```python
data.rename(columns=rename_cols, inplace=True)
```

## Drop First Row

```python
data = data.drop(index=0, axis=0)
```

## Replace zeros with null

```python
data.replace(0, pd.NA, inplace=True)
```

## Check rows with blanks

```python
empty_cols = list(data.columns[data.isnull().any()])
```

## Required Cols

```python
required_cols = ['Risk Assessment\nTaxonomy row', 'Project', 'Entity', 'Finding Title', 'Rating', 'Finding', 'IssueManager', 'IssueReviewer', 'Released', 'IssueWorkflowState', 'Due Date', 'Review Due Date', 'Actual Created Date', 'Recommendation Title', 'Recommendation', 'Management Response', 'RecommendationWorkflowState', 'Response Due Date', 'Due Date', 'Review Due Date', 'Revised Due Date', 'Actual Created Date', 'RecommendationBusinessContact', 'RecommendationBusinessObserver', 'RecommendationBusinessReviewer']
```

## Final: Check Empty Required Columns

```python
cols = [col for col in required_cols if col in empty_cols]
msg = 'There are no empty required columns.'
if len(cols):
    msg = f'There are {len(cols)} empty required columns: {cols}'

print(msg)
```
